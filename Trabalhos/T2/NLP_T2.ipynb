# Imports

```python
import pandas as pd
import re
import nltk
from nltk.corpus import stopwords
from nltk.corpus import wordnet as wn
```

```python
nltk.download('stopwords')
nltk.download('wordnet')
nltk.download('omw-1.4')
```

# Pre Processing

```python
link = "https://raw.githubusercontent.com/viniciusrpb/cic0269_natural_language_processing/main/corpus_tweets/twitter-2013train-A.txt"
df = pd.read_csv(link , sep='\t', encoding='utf-8', names = ['id', 'polarity', 'text'])
```

```python
text_list = df['text']
```

### Normalização

```python
stopwords_nltk = stopwords.words('english')
```

```python
from nltk.stem.porter import *
stemmer = PorterStemmer()
```

```python
text_list_new = []
thisset = set()
emojiDict = {':)' : 'happy', '(:' : 'happy', ':(': 'sad', '):': 'sad', '(;' : 'wink', ';)' : 'wink', ':D' : 'happy', 'D:': 'sad'}
lemmatizer = nltk.stem.WordNetLemmatizer()
wn_lemmas = set(wn.all_lemma_names())

for t in text_list:
    # substituindo os emojis por palavras
    for word, initial in emojiDict.items():
      t = t.replace(word.lower(), initial)
    
    # regex para tirar caracteres, numeros, upcode e links htpp
    words = re.sub(r'[\s!%*~\^´`=+<\[\]?&$:;@#.0-9()\/\"\'_-]+|[\s]*[\\]u[a-z0-9]{4}[\s]*|[\s]*http[^\s]+[\s]*', " ", t.lower())
    words = re.sub(r'\brt\b', "", words)
    words = re.sub(r'[\\]+', " ", words)
    words = words.strip()
    
    phrase = ""
    words = words.split()

    # tirando stopwords
    # colocando apenas as palavras que estao na gramatica wordnet
    # fazendo stematizacao
    for x in words:
        if x not in stopwords_nltk and x in wn_lemmas:
            x = stemmer.stem(x)
            thisset.add(x)
            phrase = phrase + " " + x
        

    text_list_new.append(phrase.strip())
   
```

```python
df['text'] = text_list_new

df #Impressão do Dataframe de tweets
```
